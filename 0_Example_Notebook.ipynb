##### Polynomial equations are defined as any equation that has:
1. Terms with only Positive exponents
2. Uess only addtion, subtraction, and multiplication

##### Let's take a look at some simple examples in the following form:

\\( y = mx + b \\)

***

### Define a function to calculate the results of the linear equation

```python
def poly_equ(m, equ_order, x, b):
    return (m * (x ** equ_order)) + b
```

### Import Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Setup Matplotlib for Jupyter Notebooks

```python
%matplotlib inline
#plt.rcParams['figure.figsize'] = 15, 10
```

### Run Polynomial Equations and Show Results

```python
m = 1
b = 0
df = pd.DataFrame(list(range(-20, 20, 1)), columns=list('x'))
df['y'] = df.apply(lambda row: poly_equ(m, 1, row.x, b), axis=1)
df['y2'] = df.apply(lambda row: poly_equ(m, 2, row.x, b), axis=1)
df['y3'] = df.apply(lambda row: poly_equ(m, 3, row.x, b), axis=1)
df['y4'] = df.apply(lambda row: poly_equ(m, 4, row.x, b), axis=1)
df['y5'] = df.apply(lambda row: poly_equ(m, 5, row.x, b), axis=1)
```

```python
# Draw Plots
plt.figure(1)
plt.title("1st Order")
plt.plot(df.x, df.y)

plt.figure(2)
plt.title("2nd Order")
plt.plot(df.x, df.y2)

plt.figure(3)
plt.title("3rd Order")
plt.plot(df.x, df.y3)

plt.figure(4)
plt.title("4th Order")
plt.plot(df.x, df.y4)

plt.figure(5)
plt.title("5th Order")
plt.plot(df.x, df.y5)
```
