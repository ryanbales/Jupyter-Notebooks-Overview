### Load the R Plotting Package

```r
library(ggplot2)
```

### Load the Diamonds Dataset and Display a Summary View

```r
data(diamonds)
summary(diamonds)
```

### Let's take a look at the Dataset

```r
typeof(diamonds)
```

```r
head(diamonds)
```

### Let's take a look at some features of our data

```r
ggplot(data=diamonds) + 
geom_histogram(binwidth=500, aes(x=diamonds$price)) + 
ggtitle("Diamond Price Distribution") + xlab("Price (USD)") + ylab("Frequency")
```

### Let's Review the Help Docs for the AES functionality of ggplot2

```r
help(aes)
```

### Let's Take a Closer Look at the Scatter Plots based on the 4Cs

```r
ggplot(diamonds, aes(x=carat, y=price, color=cut)) + 
geom_point() +
ggtitle("Diamond Scatter Plot - CUT") + xlab("Carat") + ylab("Price")
```

```r
ggplot(diamonds, aes(x=carat, y=price, color=clarity)) +
geom_point() + 
ggtitle("Diamond Scatter Plot - CLARITY") + xlab("Carat") + ylab("Price")
```

```r
ggplot(diamonds, aes(x=carat, y=price, color=color)) +
geom_point() +
ggtitle("Diamond Scatter Plot - COLOR") + xlab("Carat") + ylab("Price")
```

### Price Distribution Historgrams by the 4Cs

```r
ggplot(diamonds, aes(x=price, fill=cut)) + 
geom_histogram(binwidth=500, aes(x=diamonds$price)) + 
ggtitle("Diamond Price Distribution") + xlab("Price (USD)") + ylab("Frequency")
```

```r
ggplot(diamonds, aes(x=price, fill=clarity)) + 
geom_histogram(binwidth=500, aes(x=diamonds$price)) + 
ggtitle("Diamond Price Distribution") + xlab("Price (USD)") + ylab("Frequency")
```

```r
ggplot(diamonds, aes(x=price, fill=color)) + 
geom_histogram(binwidth=500, aes(x=diamonds$price)) + 
ggtitle("Diamond Price Distribution") + xlab("Price (USD)") + ylab("Frequency")
```
